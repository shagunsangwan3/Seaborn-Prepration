# Box Plot in Seaborn: Key Parameters and Components

A box plot (or box-and-whisker plot) in Seaborn is a powerful visualization tool used to summarize the distribution of a dataset, highlighting its central tendency, variability, and potential outliers.

# Key Components of a Box Plot
- Box: Represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3).
- Median Line: A horizontal line inside the box indicating the median (Q2) of the data.
- Whiskers: Extend from the box to the smallest and largest values within 1.5 times the IQR.
- Outliers: Data points outside the whiskers, plotted as individual points.
# Key Parameters in Seaborn's boxplot()
Here are some commonly used parameters:

- x / y: Variables to plot on the x-axis or y-axis.
- data: The dataset (e.g., a Pandas DataFrame) containing the variables.
- hue: Categorical variable to group data by color.
- palette: Color palette for the plot.
- orient: Orientation of the plot ('v' for vertical, 'h' for horizontal).
- width: Width of the boxes.
- showmeans: Whether to display the mean as a point.
- linewidth: Width of the box and whisker lines.
- dodge: Whether to separate boxes when using the hue parameter.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",hue="time",data=df)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",hue="time",data=df,order=["Sun","Sat","Fri","Thur"],palette="cool",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"r"},linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",hue="time",data=df,order=["Sun","Sat","Fri","Thur"],palette="cool",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"r"},linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=df,palette="cool",showmeans=True,
            meanprops={"marker":"+","markeredgecolor":"r"},linewidth=3,orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y=df["total_bill"])
plt.show()
```
