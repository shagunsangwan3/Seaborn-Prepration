# Seaborn Pair Plot: Key Parameters and Components

A pair plot in Seaborn is a powerful tool for visualizing pairwise relationships in a dataset. It creates a grid of plots, showing scatter plots for relationships between variables and histograms (or KDE plots) for individual variable distributions.

# Key Parameters
- data: The dataset (Pandas DataFrame) to visualize.
- hue: A column name to group data by color, adding a categorical distinction.
- vars: A list of specific columns to include in the pair plot.
# kind: Type of plot for diagonal elements:
- "scatter" (default): Scatter plots for relationships.
- "kde": Kernel Density Estimation plots for distributions.
- palette: Color palette for the hue variable.
- height: Size of each subplot (default is 2.5).
- diag_kind: Type of plot on the diagonal:
- "auto" (default): Automatically selects based on data type.
- "hist": Histogram.
- "kde": Kernel Density Estimation.
- corner: If True, only plots the lower triangle of the grid.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue="sex",x_vars=["total_bill","tip"],hue_order=["Female","Male"],markers=["*","<"])
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="cool",kind="kde", diag_kind="hist") # Scatter , kde , hist , reg
plt.show()
```
