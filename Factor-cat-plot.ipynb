# Factor Plot and Cat plot in Seaborn

A Factor Plot in Seaborn is a versatile tool for visualizing categorical data. It allows you to create various types of plots (e.g., point, bar, box, violin, etc.) by specifying the kind parameter. Although the factorplot() function has been deprecated since Seaborn v0.9.0, its functionality is now available through the catplot() function.

Here’s a breakdown of its key parameters and components:

# Key Parameters
- data:
The dataset (Pandas DataFrame) containing the variables to be plotted.

- x / y / hue:

- x: Categorical variable for the x-axis.
- y: Numerical variable for the y-axis.
- hue: Categorical variable for color differentiation.
- kind: Specifies the type of plot to create. Options include:
    - "point" (default)
    - "bar"
    - "box"
    - "violin"
    - "strip"
    - col / row:
# Variables to create subplots across columns or rows.

- palette:
Color palette for the plot.

- aspect:
Aspect ratio of each facet (width/height).

- height:
Height of each facet in inches.

# Components
- Axes:
Each subplot is an individual facet, allowing for multi-dimensional visualization.

- Facets:
Subplots arranged by col and row parameters.

- Legend:
Automatically generated when the hue parameter is used.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
sns.catplot(x="size", y="tip", data=df, kind="point", markers="o", linestyles="-")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,kind="box")
plt.show()
```

```python
sns.catplot(x="tip",y="size",hue="sex",data=df,palette="cool")
plt.show()
```

```python
sns.catplot(x="tip",y="size",hue="sex",data=df,palette="cool",kind="boxen")
plt.show()
```
